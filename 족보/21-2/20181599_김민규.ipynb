```python
#1
f = open("images.txt",'r')
data = f.readlines()
#파일을 닫아준다 close
f.close()
```

```python
data
```

```python
#2
addr = ""
for line in data:
    addr = addr + line.strip() + ","
addr1 = addr.split(".")
extension_list = []
for line1 in addr1:
    domain_split = line1.split(".")
    domain_list.append(domain_split[0])
extension_list=domain_list
extension_list
```

```python
#3
extension_lower = extension_list.lower()
```

```python
#4
gu_list = ['jpg','png']
gu_sum = 0
for i in gu_list:
    cnt = extension_lower.count(i)
    print(i,":",addr.count(i))
    gu_sum += cnt
print("="*15)
print("전체",":",gu_sum)
```
