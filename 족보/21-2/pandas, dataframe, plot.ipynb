# DataFrame

```python
import pandas as pd
```

```python
customers = pd.read_csv("customers.csv",encoding='utf8')
```

```python
customers.head()
```

## Dataframe의 Attributes
index 행의 이름
columns 열의 이름
shape 행과 열의 수
iloc 정수 위치 기반 인덱싱
loc lable 또는 조건을 이용한 인덱싱

```python
customers.index
```

```python
customers.columns
```

```python
customers.shape
```

```python
# iloc : 정수 위치 기반 인덱싱 - 행
customers.iloc[[0,1,2]]
```

```python
# iloc : 정수 위치 기반 인덱싱 - 열
customers.iloc[:,[3,4,5]]
```

```python
# iloc : 정수 위치 기반 인덱싱 - 행 & 열
customers.iloc[[0,1,2],[3,4,5]]
```

```python
# loc : lable 또는 조건을 이용한 인덱싱
df_USA = customers.loc[customers['country'] == 'USA']
#df_USA = customers.loc[customers.country. == 'USA']
df_USA.shape
```

```python
df_USA.head()
```

```python
# ex) 일본 사람은 몇 명 있나요?
df_Japan = customers.loc[customers.country == 'Japan']
df_Japan.shape[0]
```

## DataFrame의 Methods

```python
# head
# tail
# describe() 변수(열)별 기술 통계량
# to_clipboard : 클립보드에 복사
# to_csv : csv 파일로 저장
# to_excel : excel 파일로 저장
```

```python
df_USA.describe()
```

```python
# include='all' 추가시 고유값 개수(unique), 최빈값(top), 빈도수(freq) 추가한다.
df_USA.describe(include = 'all')
```

```python
df = pd.read_csv("airquality_20210607.csv",encoding='utf8')
df.head()
```

```python
df.shape
```

```python
df = df.sort_values(by=['YMDH'])
df.head()
```

```python
#rot x축서식 각도 조절
ax = df.plot.line(x="YMDH",y="PM10",rot=45)
```

```python
ax = df.plot.line(x="YMDH",y=["PM10","PM2.5"],rot=45)
```

```python
ax = df.plot.line(x="YMDH",rot=45)
```

```python
ax = df.plot.line(x="YMDH",rot=45,subplots=True,figsize=(6,12))
```

```python
# 산점도 - 변수 2개
ax = df.plot.scatter(x="PM10",y='PM2.5',c='darkblue')
```

```python
# 산점도 - 모든 변수
ax = pd.plotting.scatter_matrix(df,figsize=(12,12))
```
