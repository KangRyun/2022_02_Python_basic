```python
import pandas as pd
from matplotlib import pyplot as plt
### 그래프에 굴림 글꼴 사용
plt.rcParams['font.family'] = 'Gulim'
```

```python
data_file = easygui.fileopenbox()
```

```python
df_air = pd.read_csv(data_file, encoding='utf8')
df_air.head
df_air.shape
```

```python
#데이터 정렬(날짜별로)
df_air = df_air.sort_values(by=['YMDH'])
df_air.head
```

```python
df_air.plot.line(x='YMDH',y = 'PM10', rot = 45) #rot45도로
```

```python
#변수 두개를 선그래프로
df_air.plot.line(x='YMDH',y = ['PM10','PM2.5'], rot = 45) #rot45도로
```

```python
#모든 변수
df_air.plot.line(x='YMDH', rot = 45) #rot45도로
```

```python
#모든변수(subplot)
df_air.plot.line(x='YMDH', rot = 45, subplots = True, figsize = (6,12)) #rot45도로
```

```python
# 산점도 - 변수 2개
df_air.plot.scatter(x = 'PM10', y = 'PM2.5', c = 'darkblue')
```

```python
###산점도 모든변수
pd.plotting.scatter_matrix(df_air, figsize=(12, 12))
```
